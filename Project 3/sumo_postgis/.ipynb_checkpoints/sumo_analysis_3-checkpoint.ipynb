# Analiza podataka dobijenih simulacijom u SUMO-GUI okruzenju
## Import svih potrebnih biblioteka za analizu

```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine

import movingpandas as mpd
from shapely.geometry import Point
import folium
import numpy as np
```

## Uspostavljanje konekcije sa PostGIS bazom podataka

```python
engine = create_engine("postgresql://postgres:Petsre971$$$@localhost:5432/sumo_db")
```

## Uzimanje podataka ( vozila ) iz PostGIS baze podataka 

```python
fcd = gpd.read_postgis(
    """
    SELECT *
    FROM fcd_tracks
    """,
    engine,
    geom_col="geom"
)

#set_crs(4326, allow_override=True) sam napisao jer sam greskom oznacio geom podatke u POSTGIS pri importovanju podataka da su EPSG:3857
#umesto EPSG:4326 pa ispravljam gresku naknadno, plus konvertujem u EPSG:3857 jer je bolje za analizu u ovom slucaju
fcd = fcd.set_crs(4326, allow_override=True).to_crs(3857)

#ispravljaju se i x/y kolone tako da budu u ispravnom koodrinatnom sistemu
fcd["x"] = fcd.geometry.x
fcd["y"] = fcd.geometry.y

print("FCD shape:", fcd.shape)
print("\nKolone:", fcd.columns)
print("\nPrve 5 kolone:")
display(fcd.head())
```

## konverzija formata vremena iz sekundi u datetime, jer movingpandas radi sa datetime formatom

```python
import datetime

base_time = pd.Timestamp("2026-01-01")

fcd["t"] = base_time + pd.to_timedelta(fcd["time"], unit="s")

#sada su svi vremenski podaci konvertovani iz sekunda u 2026-01-01 00:00:00 + offset (tj broj sekunda).

print("Primer vremenske kolone:")
display(fcd[["time", "t"]].head())

print("Vremenski opseg:")
print(fcd["t"].min(), "→", fcd["t"].max())
```

## kreiranje trajektorija

```python
fcd_trips = mpd.TrajectoryCollection(
    fcd,
    traj_id_col="vehicle_id",
    t="t",
    x="x",
    y="y"
)

print("Broj trajektorija:")
print(len(fcd_trips))

print("\nPrimeri ID trajektorija:")
print(fcd_trips.trajectories[:5])
```

# Analiza 3 : Trajektorija određenog vozila u datom periodu

```python
#odredjivanje vremenskog intervala za koji se gleda trajektorija ( gleda se gde je sve vozilo prošlo od x vremena do y vremena )
start_time = pd.Timestamp("2026-01-01 00:00:00")
end_time   = pd.Timestamp("2026-01-01 00:02:00")

#uzima se prva trajektorija iz liste
traj = fcd_trips.trajectories[0]
print("ID trajektorije:", traj.id)

#pretvara se u GeoDataFrame
traj_points = traj.to_point_gdf()

#vrši se filtriranje po vremenu gde se koristi indeks jer je indeks zapravo vreme, tj vreme kada je neka tačka bila na nekoj koordinati.
traj_points = traj_points[
    (traj_points.index >= start_time) &
    (traj_points.index <= end_time)
].copy()

print("Tačke nakon filtriranja po vremenu:", len(traj_points))

#ovo je dodato jer može da se desi da ne postoji nijedna tačka u tom intervalu
if len(traj_points) == 0:
    raise ValueError("Ne postoje tačke u ovom vremenskom intervalu")

#pošto folium koristi 4326 sistem, moramo da konvertujemo koordinate naših tačaka trajektorije u 4326 sistem kako bismo ih predstavili na mapi
traj_points = traj_points.to_crs(4326)

print("tačke:", len(traj_points))

#konstruišemo koordinate
coords = [(p.y, p.x) for p in traj_points.geometry]

#kreiramo mapu pri čemu je coords[0] početna koordinata gde će se mapa renderovati
m = folium.Map(location=coords[0], zoom_start=14)

#crta se trajektorija koja je ustvari samo jedna polilinija tj. skup od više linija koje se predstavljaju kao jedan objekat
folium.PolyLine(
    coords,
    color="red",
    weight=5,
    opacity=0.8,
    popup=f"Trajectory: {traj.id}"
).add_to(m)

#uzimamo prvu i poslednju tačku naše trajektorije kako bismo označili početak i kraj trajektorije
start_geom = traj_points.geometry.iloc[0]
end_geom = traj_points.geometry.iloc[-1]

start_time = traj_points.index[0]
end_time = traj_points.index[-1]

#kreiramo markere za početnu i krajnju tačku trajektorije
folium.Marker(
    [start_geom.y, start_geom.x],
    popup=f"START<br>{start_time}",
    tooltip="Start"
).add_to(m)

folium.Marker(
    [end_geom.y, end_geom.x],
    popup=f"END<br>{end_time}",
    tooltip="End"
).add_to(m)

#crtamo mapu
m
```
